```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Terror = pd.read_csv("C:/Users/91727/Desktop/globalterrorismdb_0718dist.csv",on_bad_lines ='skip',encoding='latin')
pd.set_option("display.max_columns",500)
```

```python
#extracting columns
cols=['iyear','imonth','iday','country_txt','city','latitude','longitude','location','attacktype1_txt','targetype1_txt','targsubtype1_txt',
    'target1','gname','motive','weaptype1_txt','dbsource','region_txt','nkill','nwound','natlty1_txt','weapdetail']
```

```python
Terror_clean = pd.DataFrame(data = Terror,columns = cols)
```

```python
Terror.head()
```

```python
Terror_clean.head()
```

```python
Terror_clean.shape
```

```python
Terror_clean.isnull().sum()
```

```python
Terror_clean.rename(columns={'iyear':'year',
                            'imonth':'month',
                            'iday':'day',
                            'country_txt':'country name',
                            'attacktype1_txt':'attacktype',
                            'targetype1_txt':'targettype',
                            'targetsubtype1_txt':'targetsubtype',
                            'target1':'target',
                            'gname':'group name',
                            'weaptype1_txt':'weaptype',
                            'dbsource':'source',
                            'region_txt':'region',
                            'nkill':'killed',
                            'nwound':'wounded',
                            'weapdetail':'weapon_detail',
                            'natlty1_txt':'nationality',
                            'targsubtype1_txt':'targsubtype1'},inplace = True)
```

```python
Terror_clean.head(2)
```

```python
pd.set_option('display.max_columns',500)
```

```python
Terror_clean['casualities'] = Terror_clean['wounded']+Terror_clean['killed']
```

```python
Terror_clean.head(5)
```

```python
nationality_top = Terror_clean[Terror_clean['nationality'] != 'Unknown']
```

```python
nationality_type = nationality_top['nationality'].value_counts().reset_index()
```

```python
nationality_type.rename(columns = {"index":'Nationality','nationality':'Counts'}, inplace=True)
nationality_type
```

```python
Terror_clean['nationality'].unique()
```

```python
terr= Terror_clean[Terror_clean['group name'] != 'Unknown']
Terror_Group_Org=terr['group name'].value_counts().reset_index()
Terror_Group_Org.rename(columns={"index":'Group_name','group name':'Counts'},inplace=True)
Terror_Group_Org
```

```python
 city1= Terror_clean[Terror_clean['city'] != 'Unknown']
```

```python
city_cln=city1['city'].value_counts().reset_index()
city_cln.rename(columns={"index":'City_name','city':'Counts'},inplace=True)
city_cln
```

```python
wp=Terror_clean[Terror_clean['weaptype'] != 'Unknown']
weaptype_cln=wp['weaptype'].value_counts().reset_index()
weaptype_cln.rename(columns={"index":'weapon_type','weaptype':'Counts'},inplace=True)
weaptype_cln
```

```python
Terror_clean.head(1)
```

```python
f, ax = plt.subplots(figsize=(15, 10))
ax = sns.barplot(x="Counts", y="Nationality", data=nationality_type[:10],
                 palette="viridis").set_title('Nations Suffered Most of the TARGET Attacks')
```

```python
f, ax = plt.subplots(figsize=(15, 10))
ax = sns.barplot(x="Counts", y="Group_name", data=Terror_Group_Org[:10],
                 palette="Reds").set_title('Groups Responsible Behind the Terror Attacks')
```

```python
f, ax = plt.subplots(figsize=(15, 10))
ax = sns.barplot(x="Counts", y="City_name", data=city_cln[:15],
                 palette="Blues").set_title('Top 15 Cities Suffered Terror attacks')
```

```python
f, ax = plt.subplots(figsize=(15, 10))
ax = sns.barplot(x="Counts", y="weapon_type", data=weaptype_cln[:7]).set_title('Weapons Used in attack')
```

```python
 plt.subplots(figsize=(30, 15), dpi = 60)
 sns.countplot( y="attacktype", data=Terror_clean,palette = 'inferno',order= Terror_clean['attacktype'].value_counts().index)
 plt.yticks(size = 25)
 plt.xticks(size = 25)
plt.title("Attack type By terrorists",size = 20)
plt.show()
```

```python
plt.subplots(figsize=(15,6), dpi = 100)
sns.countplot( x=Terror_clean['targsubtype1'],order= Terror_clean['targsubtype1'].value_counts().index)
plt.xticks(rotation = 90)
plt.title("Target Type")
plt.show()
```

```python
plt.subplots(figsize=(15,30), dpi = 120)
sns.histplot(data = Terror_clean,
            y = 'targsubtype1',
            bins = 15,
            kde = True)
plt.xlabel('Order Counts')
plt.title('Target Type Attacked')
plt.show()
```
